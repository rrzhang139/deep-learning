# Introduction

> A goal of statistical language modeling is to learn the joint probability function of sequences of words in a language. This is intrinsically difficult because of the curse of dimensionality: a word sequence on which the model will be tested is likely to be different from all the word sequences seen during training.

> by concatenating very short overlapping sequences seen in the training set. We propose to fight the curse of dimensionality by learning a distributed representation for words which allows each training sentence to inform the model about an exponential number of semantically neighboring sentences.

> Generalization is obtained because a sequence of words that has never been seen before gets high probability if it is made of words that are similar

First convert sequences into representation so they can be compared. So for example we have a word sequence "quick fox" and "quick brown fox" we can convert them into vectors and then compare them.

$\hat{P}\left(w_1^T\right)=\prod_{t=1}^T \hat{P}\left(w_t \mid w_1^{t-1}\right)$

We can calculate the conditional probability of a next word given previous, and this will inform us what to choose

=======

1. associate with each word in the vocabulary a distributed word feature vector (a realvalued vector in \(\mathbb{R}^m\) ),
2. express the joint probability function of word sequences in terms of the feature vectors of these words in the sequence, and
3. learn simultaneously the word feature vectors and the parameters of that probability function.

Recall that we are finding a joint probability of many words. And each word has a probability function with parameters that are shaping how likely the word is to appear. We vectorize the words

Generalization is about transferring probability mass from one sequence to another and the way to do that is to make the sequences similar in the vector space.

Using Neural nets in highly dimensional, discrete distributions are useful for joint probability function learning. It can be decmposed into a product of conditional probabilities.

> $\hat{P}\left(Z_1=z_1, \cdots, Z_n=z_n\right)=\prod_i \hat{P}\left(Z_i=z_i \mid g_i\left(Z_{i-1}=z_{i-1}, Z_{i-2}=z_{i-2}, \cdots, Z_1=z_1\right)\right)$

> In contrast, here we push this idea to a large scale, and concentrate on learning a statistical model of the distribution of word sequences, rather than learning the role of words in a sentence.

This is a huge difference, because previous methods have been studying the probability of words given n words before it. We are focusing on generalization of many sequences in larger scale, using vectorization and large scale distributed compute.
This means that their approach concentrates on capturing the broader patterns and regularities in the distribution of word sequences across a large corpus of text
